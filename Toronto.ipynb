# Extracting data from the wikipedia page

The table from the wikipedia page is converted into a csv file using the link:
http://wikitable2csv.ggor.de/
This is a javascript based parser that is much more efficient for tables than BeautifulSoup

Loading the csv file into a dataframe

```python
import pandas as pd
df = pd.read_csv('Tordata.csv')
df.head()
```

# Data wrangling

Dropping all rows where the Borough is Not assigned

```python
df = df.drop(df[df.Borough == 'Not assigned'].index)
df.shape
```

Assigning the neighbourhood as the Borough if neighbourhood is not assigned

```python
for i in range(212):
    if df.iloc[i][2] == 'Not assigned':
        df.iloc[i][2] = df.iloc[i][1]
df.head(10)
```

For 2 successive rows, combining the neighbourhoods, if the postcode is same

```python
for i in range(212-1):
    if df.iloc[i][0] == df.iloc[i+1][0]:
        df.iloc[i][2] = df.iloc[i][2] + ' ,' + df.iloc[i+1][2]
        df.iloc[i+1][2] = 'Cancel'
df = df.drop(df[df.Neighbourhood == 'Cancel'].index)
df.head(10)
```
